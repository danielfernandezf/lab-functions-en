## Arbitrary Arguments ( *argss and **kwargs)

*args arbitrary positional arguments, *parametername, the cunction will recieve a *tuple* of arguments

```python
def print_all_foods(*foods):
    print("My favorite foods are: ")
    for food in foods:
        print(f"-{food}")

print_all_foods("Pizza", "Pasta", "Ice Cream")
```

```python
print_all_foods("Steak")
```

## **kwargs

**parametername, the function will recieve a dictionary of *keyword* arguments

```python
def build_profile(first,last,**user_info):
    #build a dictionary containing everything we know about a user
    user_info['first_name'] = first
    user_info['last_name'] = last
    return user_info

user_profile = build_profile(
    'albert', 'enstein',
    location = 'princeton',
    field = 'physics'
)
```

```python
user_profile
```

## Challenge 1

Write a function that takes a list of numbers and returns a new list containing only the even numbers from the original list.

```python
def evens_only(list_nums):
    evens = []
    for i in list_nums:
        if i % 2 == 0:
            evens.append(i)
    return evens
```

```python
evens_only([1, 2, 3, 4, 5, 6])
```

The difference between print() and return:

print() displays a value on the screen, while return sends a value back from a function and stops the function


```python
def password_validator(password):
    if len(password)> 8:
        print('BEFORE RETURN')
        return "ok"
        print("AFTER RETURN")
    if "-" in password:
        print('BEFORE RETURN')
        return "ok2"
        print('AFTER RERTURN')
```

```python
password_validator("test-8characters")
```

```python
password_validator("exam-2")
```

## Variable Scope

outer to inner: variables defined outside of a function are accessible within the function 

inner to outer: variables defined inside a function are not accessible outside the function

```python
def addition(a, b):
    return a+b
a=10

addition(a, 10)
```

```python
def addition(a, b):
    d = a+b
    return d
addition(10,10)
```

```python
print(d)
```

```python
## Lambdas

lambda <param list>:<return expression>

<param lit> is the list of parameters the lambda function takes

<return expression> defines what the lambda returns

Good for passing a simple function as an argument to another function
```

```python
add = lambda x, y: x + y
```

```python
result = add(3, 5)
result
```

```python
square = lambda x: x**2
result = square(4)
result
```

```python
is_even = lambda x: x%2 == 0
result = is_even(6)
result
```

```python
get_last_char = lambda s: s[-1]
result = get_last_char("Hello!")
result
```

They can also be used as arguments in calls to other cfunctions sorted() filter() map()

sorted_list = sorted(iterable, key = lambda item: custom_sort_logic(item))

iterable - collection you want sorted

custom_sort_logic(item) is the lambda that defines how items should be compared and sorted

```python
students = [
    {"name": "Alice", "age": 25},
    {"name": "Bob", "age": 60},
    {"name": "Alphonso", "age": 40}
]   
```

```python
# sorted_list = sorted(iterable, key = lambda item: custom_sort_logic(item)) reverse false por defecto, para descendente true
sorted_list = sorted(students, key = lambda x: x["age"], reverse = True)
```

## Recursion

Recursive function calls itself, refers to itself, loops back on itself

```python
# charasteristic of factorial number: n! = n * (n-1)!

def factorial_recursive(n):
    print(n, "!")
    if n <=0:
        return 1
    else:
        local_factorial = n*factorial_recursive(n-1)
        print("Factorial for " + str(n) + "! is " + str(local_factorial))
        return local_factorial

factorial_recursive(5)
```

```python
def user():
    value = input("Please enter a value less than 5:")
    while int(value)> 5:
        value = input("Please enter a value LESS THAN 5!!:")
    return int(value)
```

```python
user()
```

```python
def user():
    value = int(input("Please enter a value less than 5:"))
    if value > 5:
        user()
    return value
```

```python
user()
```
